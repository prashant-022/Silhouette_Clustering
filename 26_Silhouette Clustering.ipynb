# Performance Metrics Clustering - Silhouette Coefficient

```python
## Importing dataset and required libraries:
from sklearn .datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples,silhouette_score

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

```python
## Generating sample dataset using make_blobs:
## In this particular example, one cluster is distinct and other 3 are closed together

X,y = make_blobs(n_samples=1000,
                 n_features=2,
                centers=4,
                cluster_std=1,
                center_box=(-10.0,10.0),
                shuffle=True,
                random_state=1)    ## For reproducibility

range_n_cluster = [2,3,4,5,6]
```

```python
import pandas as pd
df_X=pd.DataFrame(X)
df_X
```

```python
df_y=pd.DataFrame(y)
df_y
```

```python
df=pd.concat([df_X,df_y], axis=1)
df.head()
```

```python
## code to optimize the number of clusters using elbow method:
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=0, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title("The Elbow Method")
plt.xlabel("K-value")
plt.ylabel("WCSS")
plt.show()
```

```python
## Fitting the cluster using Kmeans:
clusterer = KMeans(n_clusters=4,random_state=10,n_init='auto')
cluster_label = clusterer.fit_predict(X)
print(cluster_label)
```

```python
for n_clusters in range_n_cluster:
    ## creating a subplot with 1 row and 2 columns:
    fig,(ax1,ax2) = plt.subplots(1,2)
    fig.set_size_inches(18,7)
    
    ## The 1st subplot is Silhouette plot, ranges from -1 to +1. But here, ranges from -0.1 to +1:
    ax1.set_xlim([-0.1,1])
    
    ## The (n_clusters + 1)*10 is for insertinf blank spaces b/w Silhouette plots of 
    ## individual clusters to demarcate them clearly
    ax1.set_ylim([0, len(X) + (n_clusters+1)*10])
    
    ## Initialize the clusterer with n_clusters value and a random generator seed of 10 for reproducibilty
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_label = clusterer.fit_predict(X)
    
    ## The silhouette_score gives the average value for all the samples.
    ## This gives the perspective into the density and separation of the formed clusters
    silhouette_avg = silhouette_score(X,cluster_label)
    print("For n_cluster = ", n_clusters, "The average Silhouette Score is = ",silhouette_avg)
    
    ## Compute the silhouette score for each sample:
    sample_silhouette_values = silhouette_samples(X,cluster_label)
    
    y_lower = 10
    for i in range(n_clusters):
        
        ## Aggregate the silhouette scores for samples belonging to cluster i and sort them
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_label==1]
        ith_cluster_silhouette_values.sort()
        
        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i
        
        color = cm.nipy_spectral(float(i)/n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                         0, ith_cluster_silhouette_values,
                         facecolor = color, edgecolor = color, alpha=0.7)
        
        ## Label the silhouette with their cluster numbers at the middle:
        ax1.text(-0.05, y_lower + 0.5*size_cluster_i, str(i))
        
        ## Computing the new y_lower value for the next plot:
        y_lower = y_upper+10
        
    ax1.set_title("The silhouette plots for various clusters")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster Label")
    
    ## The vertical line for average silhouette score of all the variables:
    ax1.axvline(x=silhouette_avg, color='red', linestyle ="--")
    
    ax1.set_yticks([])    ## Clear the y labels/ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
    
    
    ## 2nd plot showing the actual clusters formed:
    colors = cm.nipy_spectral(cluster_label.astype(float)/n_clusters)
    ax2.scatter(X[:, 0], X[:, 1], marker = ".", s=30, lw=0, alpha=0.7, c=colors,edgecolor='k')
    
    ## labeling the clusters:
    centers = clusterer.cluster_centers_
    ## Draw white circles at cluster centres:
    ax2.scatter(centers[:,0],centers[:,1],marker='o',c='white',alpha=1,s=200,edgecolor='k')
    
    for i,c in enumerate(centers):
        ax2.scatter(c[0],c[1], marker = '$%d$' %i, alpha=1, s=50, edgecolor='k')
        
    ax2.set_title('The visualization of clustered data')
    ax2.set_xlabel('Feature space for the first feature')
    ax2.set_ylabel('Feature space for the second feature')
    
    plt.suptitle(("Silhouette Analysis for KMeans clustering on sample data "
                "with n_clusters = %d" %n_clusters), fontsize=14, fontweight = 'bold')
    
    plt.show()
```
